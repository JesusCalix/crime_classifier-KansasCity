<h3> Race variable cleaning</h3>

```python
print(crimes[['Race']].groupby(['Race'])['Race'].count().reset_index(name='count').sort_values(['count'], ascending=False))
print('The null values are: ', crimes['Race'].isna().sum())
crimes['Race'].fillna('U', inplace=True)
print(crimes[['Race']].groupby(['Race'])['Race'].count().reset_index(name='count').sort_values(['count'], ascending=False))
# B – Black or African American; I – American Indian or Alaska Native ; U - Unknown; W – White; Blank - Not Provided

crimes.loc[(crimes['Race'] == 'A') | (crimes['Race'] == 'P'), 'Race'] = 'U'
print(crimes[['Race']].groupby(['Race'])['Race'].count().reset_index(name='count').sort_values(['count'], ascending=False))

crimes.loc[(crimes['Race'] == 'B'), 'Race'] = 'Black'
crimes.loc[(crimes['Race'] == 'W'), 'Race'] = 'White'
crimes.loc[(crimes['Race'] == 'U'), 'Race'] = 'Unknown'
crimes.loc[(crimes['Race'] == 'I'), 'Race'] = 'American Indian'

print(crimes[['Race']].groupby(['Race'])['Race'].count().reset_index(name='count').sort_values(['count'], ascending=False))
print('The null values are: ', crimes['Race'].isna().sum())
```

```python
def remove_outlier(df_col):
    sorted(df_col)
    q1, q3 = df_col.quantile([0.25,0.75])
    iqr = q3 - q1
    lower_range = q1 - (1.5*iqr)
    upper_range = q3 + (1.5*iqr)
    return lower_range, upper_range

lower_range, upper_range = remove_outlier(crimes['Age'])
crimes['Age'] = np.where(crimes['Age'] > upper_range, upper_range, crimes['Age'])
crimes['Age'] = np.where(crimes['Age'] < lower_range, lower_range, crimes['Age'])
```

```python
crimes['Offense_Cleaned'] = crimes['Offense']

crimes.loc[(crimes['Offense_Cleaned'].str.contains("assault",case=False)), 'Offense_Cleaned'] = 'Assault'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("assa",case=False)), 'Offense_Cleaned'] = 'Assault'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("property damage",case=False)), 'Offense_Cleaned'] = 'Vandalism'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("property",case=False)), 'Offense_Cleaned'] = 'Vandalism'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("burglary",case=False)), 'Offense_Cleaned'] = 'Burglary'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("rob",case=False)), 'Offense_Cleaned'] = 'Robbery'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("rape",case=False)), 'Offense_Cleaned'] = 'Rape'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("abuse",case=False)), 'Offense_Cleaned'] = 'Rape'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("theft",case=False)), 'Offense_Cleaned'] = 'Theft'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("stealing",case=False)), 'Offense_Cleaned'] = 'Theft'
crimes.loc[(crimes['Offense_Cleaned'].str.contains("stolen",case=False)), 'Offense_Cleaned'] = 'Theft'
# crimes.loc[(crimes['Offense_Cleaned'].str.contains("human trafficking",case=False)), 'Offense_Cleaned'] = 'Human Trafficking'
# crimes.loc[(crimes['Offense_Cleaned'].str.contains("homicide",case=False)), 'Offense_Cleaned'] = 'Homicide'
# crimes.loc[(crimes['Offense_Cleaned'].str.contains("murder",case=False)), 'Offense_Cleaned'] = 'Homicide'


print(crimes[['Offense_Cleaned']].groupby(['Offense_Cleaned'],)['Offense_Cleaned'].count().reset_index(name='count').sort_values(['count'], ascending=False))
```

```python
offenses_freq = (crimes['Offense_Cleaned'].value_counts())/crimes.shape[0]
# print(offenses_freq)
other_offenses = offenses_freq.quantile(q=0.95)
low_freq = offenses_freq[offenses_freq<=other_offenses]
crimes.loc[crimes["Offense_Cleaned"].isin(low_freq.index.tolist()),'Offense_Cleaned'] = "Other Offenses"

print(crimes[['Offense_Cleaned']].groupby(['Offense_Cleaned'],)['Offense_Cleaned'].count().reset_index(name='count').sort_values(['count'], ascending=False))
```

<h3> Transformation </h3>

```python
crimes.loc[(crimes['Firearm Used Flag'] == True), 'Firearm_Flag'] = 1
crimes.loc[(crimes['Firearm Used Flag'] == False), 'Firearm_Flag'] = 0

crimes = crimes.drop(['Firearm Used Flag'], axis=1)
print(crimes[['Firearm_Flag']].groupby(['Firearm_Flag'],)['Firearm_Flag'].count().reset_index(name='count'))
```

```python
crimes.loc[(crimes['Age'] <= 25), 'Age_Categorical'] = '0-25 years'
crimes.loc[(crimes['Age'] > 25) & (crimes['Age'] <= 50), 'Age_Categorical'] = '26-50 years'
crimes.loc[(crimes['Age'] > 50) & (crimes['Age'] <= 75), 'Age_Categorical'] = '51-75 years'
crimes.loc[(crimes['Age'] > 75) & (crimes['Age'] <= 99), 'Age_Categorical'] = '76-99 years'

# crimes = crimes.drop(['Age'], axis=1)
# crimes = crimes.rename(columns={'Age_Categorical': 'Age'})
```

```python
# print(crimes[['From_Time']].groupby(['From_Time'])['From_Time'].count().reset_index(name='count'))
crimes.loc[(crimes['Reported_Time'] == ':'), 'Reported_Time'] = '00:00'
crimes['Reported_Date_Time'] = pd.to_datetime(crimes['Reported_Date'] + ' ' + crimes['Reported_Time']+':00')
crimes.head()
```

```python
# print(crimes[['From_Time']].groupby(['From_Time'])['From_Time'].count().reset_index(name='count'))
crimes.loc[(crimes['From_Time'] == ':'), 'From_Time'] = '00:00'
crimes.loc[(crimes['From_Time'] == '24:00:00'), 'From_Time'] = '00:00'
crimes['From_Date_Time'] = pd.to_datetime(crimes['From_Date'] + ' ' + crimes['From_Time']+':00')
crimes.head()
```

```python
crimes['Reported_Date_Time:year'] = crimes['Reported_Date_Time'].dt.year
crimes['Reported_Date_Time:month'] = crimes['Reported_Date_Time'].dt.month
crimes['Reported_Date_Time:day'] = crimes['Reported_Date_Time'].dt.day
crimes['Reported_Date_Time:day_of_week'] = crimes['Reported_Date_Time'].dt.day_of_week
crimes['Reported_Date_Time:hour'] = crimes['Reported_Date_Time'].dt.hour
crimes.head()
```

```python
def day_name(day):
    if day == 0:
        return "Monday"
    elif day == 1:
        return "Tuesday"
    elif day == 2:
        return "Wednesday"
    elif day == 3:
        return "Thursday"
    elif day == 4:
        return "Friday"
    elif day == 5:
        return "Saturday"
    elif day == 6:
        return "Sunday"

crimes['Reported_Date_Time:day_of_week'] = crimes['Reported_Date_Time'].dt.day_of_week.apply(day_name)
crimes.head()
```

```python
def month_name(month):
    if month == 1:
        return "January"
    elif month == 2:
        return "February"
    elif month == 3:
        return "March"
    elif month == 4:
        return "April"
    elif month == 5:
        return "May"
    elif month == 6:
        return "June"
    elif month == 7:
        return "July"
    elif month == 8:
        return "August"
    elif month == 9:
        return "September"
    elif month == 10:
        return "October"
    elif month == 11:
        return "November"
    elif month == 12:
        return "December"
    
crimes['Reported_Date_Time:month'] = crimes['Reported_Date_Time:month'].apply(month_name)
crimes.head()
```

```python
def day_part(hour):
    if hour in [4,5]:
        return "dawn"
    elif hour in [6,7]:
        return "early morning"
    elif hour in [8,9,10]:
        return "late morning"
    elif hour in [11,12,13]:
        return "noon"
    elif hour in [14,15,16]:
        return "afternoon"
    elif hour in [17, 18,19]:
        return "evening"
    elif hour in [20, 21, 22]:
        return "night"
    elif hour in [23,24,1,2,3]:
        return "midnight"

#Run function with apply method
crimes['Reported_Date_Time:hour_part'] = crimes['Reported_Date_Time:hour'].apply(day_part)
crimes.head()
```
